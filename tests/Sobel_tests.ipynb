```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob

%matplotlib inline
```

```python
image = 'signs_vehicles_xygrad.png'
img = mpimg.imread(image)
```

```python
plt.imshow(img)
```

```python
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0,255)):
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    if orient=='x':
        grad = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = sobel_kernel)
    else:
        grad = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = sobel_kernel)
        
    grad = np.uint8(255*np.absolute(grad)/np.max(grad))
    bins = np.zeros_like(img)
    bins[(grad >= thresh[0]) & (grad <= thresh[1])] = 1
    
    return bins

def mag_thresh(img, sobel_kernel=3, mag_thresh=(0,255)):
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gradx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = sobel_kernel)
    grady = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = sobel_kernel)
    mag = np.sqrt(np.square(gradx) + np.square(grady))
    mag = np.uint8(255*mag/np.max(mag))
    bins = np.zeros_like(img)
    bins[(mag >= mag_thresh[0]) & (mag <= mag_thresh[1])] = 1
    return bins

def dir_threshold(img, sobel_kernel=3, thresh=(0,np.pi/2)):
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gradx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = sobel_kernel)
    grady = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = sobel_kernel)  
    gradx = np.absolute(gradx)
    grady = np.absolute(grady)
    
    ang = np.arctan2(grady, gradx)
    bins = np.zeros_like(img)
    bins[(ang>=thresh[0]) & (ang<=thresh[1])] = 1
    
    return bins
```

```python
res = abs_sobel_thresh(img, 'y', 3, [40,90])
plt.imshow(res, cmap='gray')
```

```python
res = mag_thresh(img, 15, [40,90])
plt.imshow(res, cmap='gray')
```

```python
res = dir_threshold(img, 15, [0.7, 1.3])
plt.imshow(res, cmap='gray')
```

```python
# Choose a Sobel kernel size
ksize = 15 # Choose a larger odd number to smooth gradient measurements

# Apply each of the thresholding functions
gradx = abs_sobel_thresh(img, orient='x', sobel_kernel=ksize, thresh=(100, 200))
grady = abs_sobel_thresh(img, orient='y', sobel_kernel=ksize, thresh=(140, 200))
mag_binary = mag_thresh(img, sobel_kernel=ksize, mag_thresh=(50, 250))
dir_binary = dir_threshold(img, sobel_kernel=ksize, thresh=(0.7, 1.3))

combined = np.zeros_like(img)
combined[((gradx == 1) & (grady == 1)) | ((mag_binary == 1) & (dir_binary == 1))] = 1
#combined[((mag_binary == 1) & (dir_binary == 1))] = 1

plt.imshow(combined, cmap='gray')
```
